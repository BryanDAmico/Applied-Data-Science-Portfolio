```python
# Importing packages
import requests  
import pandas as pd
import sklearn
import nltk
import re
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from sklearn.feature_extraction import text
from sklearn.metrics import ConfusionMatrixDisplay
from nltk.stem import WordNetLemmatizer
from sklearn import svm
import time
```

```python
## Create a new csv file to save the headlines
filename="MTGdata.csv"
MyFILE=open(filename,"w")  
## with open
### Place the column names in - write to the first row
WriteThis="Name,Text,Flavor,Color,Cost,Type,Rarity\n"
MyFILE.write(WriteThis)
MyFILE.close()
```

```python
#Getting the white cards
list_of_white_cards = []

for i in range(1,23):
    response = requests.get("https://api.scryfall.com/cards/search?unique&order&q=c=w&=lang:en&q=-t:land&page=" + str(i))
    jsontxt = response.json()
    list_of_white_cards.append(jsontxt)
    time.sleep(2)
```

```python
#Adding the white cards to the data
for page in list_of_white_cards:
    MyFILE=open(filename, "a")

    for items in page["data"]:

        if "name" in items:
            Name=items["name"]
            Name = str(Name)
            Name=Name.replace(',', '')
            Name=' '.join(Name.split())
            Name=re.sub("\n|\r", "", Name)
        else:
            Name = "NA"

        if "oracle_text" in items:
            Text=items["oracle_text"]
            Text=str(Text)      
            Text=re.sub(r'[,.;@#?!&$\-\'\(\)]+', ' ', str(Text), flags=re.IGNORECASE)
            Text=re.sub(' +', ' ', str(Text), flags=re.IGNORECASE)
            Text=re.sub(r'\"', ' ', str(Text), flags=re.IGNORECASE)
            Text=Text.replace(',', '')
            Text=' '.join(Text.split())
            Text=re.sub("\n|\r", "", Text)
        else:
            Text = "NA"

        if "flavor_text" in items:
            Flavor = items["flavor_text"]
            Flavor=str(Flavor)
            Flavor=re.sub(r'[,.;@#?!&$\-\'\(\)]+', ' ', str(Flavor), flags=re.IGNORECASE)
            Flavor=re.sub(' +', ' ', str(Flavor), flags=re.IGNORECASE)
            Flavor=re.sub(r'\"', ' ', str(Flavor), flags=re.IGNORECASE)
            Flavor=Flavor.replace(',', '')
            Flavor=' '.join(Flavor.split())
            Flavor=re.sub("\n|\r", "", Flavor)  
        else:
            Flavor = "NA"

        if "colors" in items:
            Color = items["colors"]
        else:
            Color = "NA"

        if "mana_cost" in items:
            Cost = items["mana_cost"]
        else:
            Cost = "NA"

        if "type_line" in items:
            Type = items["type_line"]
            Type=re.sub(r'[,.;@#?!&$\-\'\(\)]+', ' ', str(Type), flags=re.IGNORECASE)
            Type=re.sub(' +', ' ', str(Type), flags=re.IGNORECASE)
            Type=re.sub(r'\"', ' ', str(Type), flags=re.IGNORECASE)
            Type=Type.replace(',', '')
            Type=' '.join(Type.split())
            Type=re.sub("\n|\r", "", Type)
        else:
            Type = "NA"

        if "rarity" in items:
            Rarity = items["rarity"]
        else:
            Rarity = "NA"

        WriteThis=str(Name)+","+str(Text)+","+str(Flavor)+","+ str(Color) + "," + str(Cost) + "," + str(Type) + "," + str(Rarity) + "\n"

        #Writing the line in the csv file
        MyFILE.write(WriteThis)

    #Closing the file when the writing operation finishes
    MyFILE.close()
```
